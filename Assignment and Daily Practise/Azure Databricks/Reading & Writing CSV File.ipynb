# List of Methods to Read a CSV File in Python

## 1. Read CSV file using csv.reader

```python
import csv
rows = []
with open("/Volumes/firstworkspace/default/files/export.csv", 'r') as file:
        csvreader = csv.reader(file)
        header = next(csvreader)
        for row in csvreader:
            rows.append(row)
print(header)
print(rows)            



```

## 2. Read CSV Files in python Using .readlines()

```python
with open('/Volumes/firstworkspace/default/files/export.csv') as file:
    content = file.readlines()
header = content[:1]
rows = content[1:]
print(header)
print(rows)
```

# List of Methods to Write a CSV file in python


## 1. Write CSV file using csv.writer

```python
import csv

header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]

filename = '/Volumes/firstworkspace/default/files/students_data.csv'

with open(filename, 'w', newline="") as file:
    csvwriter = csv.writer(file)
    csvwriter.writerow(header)
    csvwriter.writerows(data)

print("CSV file saved to:", filename)
```

## 2. Write CSV Using Pandas

```python
import pandas as pd

# Data
header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]

# Create DataFrame
df = pd.DataFrame(data, columns=header)

# Save to Unity Catalog volume
df.to_csv('/Volumes/firstworkspace/default/files/stud2.csv', index=False)

print("CSV file saved to /Volumes/firstworkspace/default/files/stud2.csv")
```

```python
display(spark.read.csv("/Volumes/firstworkspace/default/files/stud2.csv", header=True))
```

## 3. Write CSV File Using csv.DictWriter


```python
import csv

# Data to write
data = [
    {'Name': 'Alex', 'M1 Score': 62, 'M2 Score': 80},
    {'Name': 'Brad', 'M1 Score': 45, 'M2 Score': 56},
    {'Name': 'Joey', 'M1 Score': 85, 'M2 Score': 98}
]

# Field names (header)
fieldnames = ['Name', 'M1 Score', 'M2 Score']

# Path to Unity Catalog volume in Databricks
filename = '/Volumes/firstworkspace/default/files/stud2.csv'

# Writing CSV file
with open(filename, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)

print(f"CSV saved to {filename}")
```
