## 8. Apply Function

```python
df['Value_len'] = df['Value'].apply(len)
```

```python
df[['Value', 'Value_len']].head()
```

## 9. Using the lambda operator

```python
# Convert Value to numeric
df['Value_num'] = df['Value'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df[['Value', 'Value_num']].head()
```
